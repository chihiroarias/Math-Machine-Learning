### 1.1 Matrices generales: `numpy.linalg.eig` y `numpy.linalg.eigvals`
Numpy ofrece funciones para calcular valores y vectores propios de una matriz. La principal de ellas es `numpy.linalg.eig`, que calcula tanto valores como vectores propios de una matriz cualquiera. Otra función es `numpy.linalg.eigvals`, que calcula solo los valores propios de una matriz. A continuación se ilustra su uso.

```python
import numpy as np

# Matriz no simétrica
A = np.array([[4, 2], [1, 3]])

# Calculamos los valores y vectores propios con eig
eigenvalues, eigenvectors = np.linalg.eig(A)
print("Eigenvalues (eig):", eigenvalues)
print("Eigenvectors (eig):")
print(eigenvectors)

# Calculamos únicamente los valores propios con eigvals
eigenvalues = np.linalg.eigvals(A)
print("Eigenvalues (eigvals):", eigenvalues)

# Comparamos los tiempos de ejecución de eig y eigvals
import timeit
print("eig:", timeit.timeit(lambda: np.linalg.eig(A), number=10000))
print("eigvals:", timeit.timeit(lambda: np.linalg.eigvals(A), number=10000))
```

### 1.2 Matrices simétricas
Numpy ofrece otra función para cálculo de valores y vectores propios:  `numpy.linalg.eigh`.   Esta función calcula valores y vectores propios de una matriz cuadrada simétrica.  El motivo para contar con una función específiica para matrices simétricas es que su estructura puede ser explotada para obtener una mayor eficiencia y precisión numérica.  También se ofrece la función `numpy.linalg.eigvalsh` que calcula solo los valores propios de una matriz simétrica.  A continuación se ilustra su uso con un ejemplo.

```python
# Matriz simétrica
B = np.array([[1, 2], [2, 3]])

# Cálculo de los valores y vectores propios con eigh
eigenvalues_sym, eigenvectors_sym = np.linalg.eigh(B)
print("Eigenvalues (eigh):", eigenvalues_sym)
print("Eigenvectors (eigh):")
print(eigenvectors_sym)

# Cálculo únicamente de los valores propios con eighvals
eigenvalues_sym = np.linalg.eigvalsh(B)
print("Eigenvalues (eighvals):", eigenvalues_sym)

# Comparamos los tiempos de ejecución
print("eigh:", timeit.timeit(lambda: np.linalg.eigh(B), number=10000))
print("eigvalsh:", timeit.timeit(lambda: np.linalg.eigvalsh(B), number=10000))
```

## 2. Métodos iterativos para calcular valores y vectores propios
Si bien existen métodos directos para calcular valores y vectores propios de una matriz, estos métodos pueden ser computacionalmente costosos y poco eficientes para matrices grandes. Por esta razón, se han desarrollado métodos iterativos que permiten calcular valores y vectores propios de una matriz de forma más eficiente. A continuación se presentan tres métodos iterativos para calcular valores y vectores propios de una matriz.

### 2.1 Método de la potencia
El método de la potencia es un método iterativo que permite calcular el valor propio dominante (el de mayor valor absoluto) de una matriz y uno de sus vectores propios asociado. El método consiste en iterar la siguiente fórmula:
1. Se elige un vector aleatorio $x_0$.
2. Se calcula $x_{k+1} = Ax_k$.
3. Se repite el paso 2 hasta que se alcance la convergencia.

Luego de un número suficiente de iteraciones, el vector $x_k$ converge a un vector propio asociado al valor propio dominante de la matriz $A$. Una vez obtenido el vector propio, se puede calcular el valor propio correspondiente mediante la fórmula $\lambda = \frac{x^T A x}{x^T x}$.  A esta expresión se le conoce como el cociente de Rayleigh de la matriz $A$ y el vector $x$.

#### Demostración de convergencia en el caso especial en que todos los valores propios son distintos en valor absoluto
Supongamos que la matriz $A$ tiene $n$ valores propios distintos en valor absoluto, es decir, $|\lambda_1| > |\lambda_2| > \ldots > |\lambda_n|$.  Notemos que esto implica que los valores propios son todos distintos entre sí, y por tanto existen $n$ vectores propios linealmente independientes (matriz diagonalizable) que forman una base de $\mathbb{R}^n$.  Sea $x_0$ un vector cualquiera.  Si lo representamos en la base de vectores propios:
$$
x_0 = \sum_{i=1}^{n} c_i v_i
$$

donde $v_i$ es el vector propio asociado al valor propio $\lambda_i$.

Al aplicar la matriz $A$ repetidamente al vector $x_0$:

$$
A x_0 = \sum_{i=1}^{n} c_i \lambda_i v_i
$$

$$
A^2 x_0 = \sum_{i=1}^{n} c_i \lambda_i^2 v_i
$$

$$
A^k x_0 = \sum_{i=1}^{n} c_i \lambda_i^k v_i
$$

Si dividimos por $\lambda_1^k$:

$$
\frac{A^k x_0}{\lambda_1^k}
=
c_1 v_1
+
\sum_{i=2}^{n} c_i
\left(\frac{\lambda_i}{\lambda_1}\right)^k v_i
$$

Como $|\lambda_1| > |\lambda_i|$ para $i > 1$, entonces

$$
\left|\frac{\lambda_i}{\lambda_1}\right| < 1
$$

y por tanto el término

$$
\left(\frac{\lambda_i}{\lambda_1}\right)^k
$$

tiende a cero cuando $k \to \infty$.

Por lo tanto:

$$
\lim_{k \to \infty}
\frac{A^k x_0}{\lambda_1^k}
=
c_1 v_1
$$

Esto implica que el vector $x_k = A^k x_0$ converge al vector propio asociado al valor propio dominante $\lambda_1$.

A su vez, el valor propio se puede calcular como:

$$
\lambda_1 =
\frac{x_k^T A x_k}{x_k^T x_k}
=
\frac{x_k^T \lambda_1 x_k}{x_k^T x_k}
=
\frac{\lambda_1 x_k^T x_k}{x_k^T x_k}
=
\lambda_1
$$

Por una cuestión de estabilidad numérica, es conveniente normalizar el vector $x_k$ en cada iteración.  Esto se logra dividiendo el vector por su norma Euclidiana.  El algoritmo del método de la potencia es el siguiente:
1. Se elige un vector aleatorio $x_0$.
2. Se normaliza el vector $x_0 = \frac{x_0}{\|x_0\|}$.
3. Se calcula $x_{k+1} = Ax_k$.
4. Se normaliza el vector $x_{k+1}= \frac{x_{k+1}}{\|x_{k+1}\|}$.
5. Se repiten los pasos 3 y 4 hasta que se alcance la convergencia.

```python
def vector_propio_dominante_iterativo(A, max_iter=1000):
    x = np.random.rand(A.shape[1])  # Vector inicial aleatorio
    for i in range(max_iter):
        x = A @ x
        x /= np.linalg.norm(x)
    l = coeficiente_rayleigh(A, x)
    return x, l

def coeficiente_rayleigh(A, x):
    lam = (x.T @ A @ x) / (x.T @ x)
    return lam

# Ejemplo de prueba
M = np.array([[4, 2], [1, 3]])
x, l = vector_propio_dominante_iterativo(M)
print("Vector propio dominante:", x)
print("Valor propio dominante:", l)

#Comparación con eig
eigenvalues, eigenvectors = np.linalg.eig(M)
print("Eigenvalues (eig):", eigenvalues)
print("Eigenvectors (eig):")
```

### 2.2 OPCIONAL Método de la potencia inversa

El método de la potencia inversa es una variante del método de la potencia que permite calcular el valor propio más pequeño (en valor absoluto) de una matriz y uno de sus vectores propios asociados. El método consiste en iterar la siguiente fórmula:
1. Se elige un vector aleatorio $x_0$.
2. Se calcula $x_{k+1} = A^{-1}x_k$.
3. Se repite el paso 2 hasta que se alcance la convergencia.

Para evitar el cálculo repetido de la inversa de la matriz $A$, típicamente se calcula $B = A^{-1}$ y se itera la fórmula $x_{k+1} = Bx_k$.

La demostración de convergencia en este caso es simple: los valores propios de la matriz $B=A^{-1}$ son los inversos de los valores propios de la matriz $A$.  Por tanto, el valor propio dominante de $AB$ es el inverso del valor propio más pequeño (en valor absoluto) de $A$.  Por lo tanto, el método de la potencia inversa converge a un vector propio asociado a este.

### 2.3 OPCIONAL Método de la potencia inversa con desplazamiento
El método de la potencia inversa con desplazamiento es una variante del método de la potencia inversa que permite calcular un valor propio cercano a un valor deseado.  
El método consiste en iterar la siguiente fórmula:
1. Se elige un vector aleatorio $x_0$.
2. Se calcula $x_{k+1} = (A - \sigma I)^{-1}x_k$.
3. Se normaliza el vector $x_{k+1}= \frac{x_{k+1}}{\|x_{k+1}\|}$.
4. Se repite el paso 2 y 3 hasta que se alcance la convergencia.

En otras palabras, el método consiste en definir $C=(A-\sigma I)^{-1}$ y aplicar el método de la potencia a la matriz $C$.  Para mostrar la convergencia, notemos que los valores propios de la matriz $C$ son los inversos de los valores propios de la matriz $A-\sigma I$.  Por tanto, el valor propio dominante de $C$ es el inverso del valor propio más pequeño, en valor absoluto, de la matriz $A-\sigma I$.  Pero los valores propios de $A-\sigma I$ son los valores propios de $A$ desplazados por $\sigma$.  Por tanto, el menor valor propio de $A-\sigma I$ es el valor propio de $A$ más cercano a $\sigma$.  Por lo tanto, el método de la potencia inversa con desplazamiento converge al valor propio más cercano a $\sigma$.

### 3.3 Implementación de PCA desde cero

En esta sección, implementaremos PCA desde cero utilizando NumPy.  

**Ejercicio 3.3.1:** Implemente la función `pca` que calcula los componentes principales de un conjunto de datos.  La función debe devolver los componentes principales y la varianza explicada por cada componente.

```python
def pca(X, k):
    # Centrar los datos
    X_demeaned = X - np.mean(X, axis=1, keepdims=True)

    # Calcular la matriz de covarianza
    S = np.cov(X_demeaned)

    # Calcular los valores y vectores propios
    eigenvalues, eigenvectors = np.linalg.eig(S)

    # Ordenar los valores y vectores propios
    sorted_indices = np.argsort(eigenvalues)[::-1]
    eigenvalues = eigenvalues[sorted_indices]
    eigenvectors = eigenvectors[:, sorted_indices]

    # Seleccionar los k primeros vectores propios
    principal_components = eigenvectors[:, :k]

    # Proyectar los datos originales a las nuevas dimensiones
    X_pca = principal_components.T @ X_demeaned

    # Calcular la varianza explicada
    variance_explained = eigenvalues[:k] / np.sum(eigenvalues)

    return X_pca, principal_components, variance_explained
    
```


**Ejercicio 3.3.2:** Compare los resultados de su función y la implementación de `sklearn.decomposition.PCA` para una matriz aleatoria de tamaño 10x1000.  Asegúrese de que los resultados sean consistentes.

```python
random_mat = np.random.rand(10, 10000)

# PCA con nuestra funcion
X_pca_custom, components_custom, variance_custom = pca(random_mat, 5)

# PCA con sklearn
pca_sk = PCA(n_components=5)
pca_sk.fit(random_mat.T)  # sklearn espera (n, d)
X_pca_sk = pca_sk.transform(random_mat.T).T  # (k x n)

# Comparar varianza explicada
print("Varianza explicada (custom):", variance_custom)
print("Varianza explicada (sklearn):", pca_sk.explained_variance_ratio_)
print("Varianzas coinciden:", np.allclose(variance_custom, pca_sk.explained_variance_ratio_))

# Comparar proyecciones (pueden diferir en signo, se compara valor absoluto)
print("Proyecciones coinciden:", np.allclose(np.abs(X_pca_custom), np.abs(X_pca_sk)))

# Comparar componentes principales (pueden diferir en signo)
print("Componentes coinciden:", np.allclose(np.abs(components_custom.T), np.abs(pca_sk.components_)))
```

```python
# Parte 1: PCA completo (784 componentes) y curva de varianza acumulada
# Usamos el pca_mnist ya fitteado con 50 componentes? No, necesitamos TODAS las 784.
# Fitteamos un PCA con todas las componentes para tener la varianza explicada completa.
pca_full = PCA(n_components=784)
pca_full.fit(X_mnist.T)

# Varianza acumulada como porcentaje
cumulative_variance = np.cumsum(pca_full.explained_variance_ratio_) * 100

plt.figure(figsize=(10, 6))
plt.plot(range(1, 785), cumulative_variance, linewidth=2)
plt.xlabel("Numero de componentes principales (k)")
plt.ylabel("Varianza explicada acumulada (%)")
plt.title("Varianza explicada acumulada vs. k componentes principales (MNIST)")
plt.grid(True, alpha=0.3)
plt.axhline(y=90, color='r', linestyle='--', label='90%')
plt.axhline(y=95, color='g', linestyle='--', label='95%')
plt.axhline(y=99, color='b', linestyle='--', label='99%')
plt.legend()
plt.tight_layout()
plt.show()

# Parte 2: Encontrar k para 90%, 95%, 99%
thresholds = [90, 95, 99]
k_values = {}
for t in thresholds:
    # El primer k donde la varianza acumulada supera el umbral
    k = np.argmax(cumulative_variance >= t) + 1  # +1 porque argmax da indice 0-based
    k_values[t] = k
    print(f"Para explicar el {t}% de la varianza se necesitan k = {k} componentes")

# Parte 3: Reconstruccion de imagenes con los k hallados
np.random.seed(42)
idx_sample = np.random.choice(X_mnist.shape[1], 5, replace=False)
ks_threshold = [k_values[90], k_values[95], k_values[99]]

mean_full = pca_full.mean_
X_centered_full = X_mnist.T - mean_full  # (n, d)

fig, axes = plt.subplots(5, len(ks_threshold) + 1, figsize=(14, 15))
fig.suptitle("Original vs Reconstrucciones (k para 90%, 95%, 99% varianza)", fontsize=16)

for row, idx in enumerate(idx_sample):
    # Imagen original
    axes[row, 0].imshow(X_mnist[:, idx].reshape(28, 28), cmap='gray')
    axes[row, 0].set_title(f"Original (y={y_mnist[idx]})")
    axes[row, 0].axis('off')

    for col, k in enumerate(ks_threshold):
        W_k = pca_full.components_[:k]  # (k, 784)
        projected = W_k @ X_centered_full[idx]  # (k,)
        reconstructed = projected @ W_k + mean_full  # (784,)
        axes[row, col + 1].imshow(reconstructed.reshape(28, 28), cmap='gray')
        axes[row, col + 1].set_title(f"k={k} ({thresholds[col]}%)")
        axes[row, col + 1].axis('off')

plt.tight_layout()
plt.show()
```

Al usar PCA, proyectamos los datos en un espacio de menor dimensión.  Pero, ¿cuál es este subespacio? ¿Cuál es el sistema de coordenadas que estamos utilizando?  ¿Tiene algún significado?  Un acercamiento a estas preguntas es pensar en las componentes principales como direcciones en el espacio original de los datos.  En este sentido, cada componente principal puede verse como un vector, o elemento, de este espacio.  En el caso del MNIST, esto quiere decir que cada componente principal puede verse como una imagen de 28x28 píxeles.
```python
# Parte 1: Visualizar las 5 primeras componentes principales de MNIST como imagenes
# Reutilizamos pca_full que ya tiene las 784 componentes fitteadas (del Problema 6)
fig, axes = plt.subplots(1, 5, figsize=(20, 4))
fig.suptitle("Primeras 5 componentes principales de MNIST", fontsize=16)

for i, ax in enumerate(axes):
    component_img = pca_full.components_[i].reshape(28, 28)
    # seismic: azul = valores negativos, blanco = cero, rojo = valores positivos
    im = ax.imshow(component_img, cmap='seismic', vmin=-component_img.max(), vmax=component_img.max())
    ax.set_title(f"PC {i+1}\n(var: {pca_full.explained_variance_ratio_[i]*100:.1f}%)")
    ax.axis('off')
    plt.colorbar(im, ax=ax, fraction=0.046)

plt.tight_layout()
plt.show()

# Interpretacion:
# - PC1 captura el contraste global (centro brillante vs bordes), diferencia digitos "gruesos" de "finos"
# - PC2 captura variacion vertical: distingue digitos que se inclinan arriba vs abajo
# - PC3 captura variacion horizontal/diagonal: separacion izquierda-derecha
# - Las componentes sucesivas capturan detalles cada vez mas finos (curvas, trazos especificos)
# - Cada componente es ortonormal a las demas: representan variaciones independientes

# Parte 2 (OPCIONAL): Repetir con olivetti_faces
from sklearn.datasets import fetch_olivetti_faces

olivetti = fetch_olivetti_faces()
X_faces = olivetti.data.T  # (4096, 400) — imagenes 64x64, 400 muestras

pca_faces = PCA(n_components=5)
pca_faces.fit(X_faces.T)

fig, axes = plt.subplots(1, 5, figsize=(20, 4))
fig.suptitle("Primeras 5 componentes principales de Olivetti Faces (eigenfaces)", fontsize=16)

for i, ax in enumerate(axes):
    face_img = pca_faces.components_[i].reshape(64, 64)
    im = ax.imshow(face_img, cmap='coolwarm', vmin=-face_img.max(), vmax=face_img.max())
    ax.set_title(f"PC {i+1}\n(var: {pca_faces.explained_variance_ratio_[i]*100:.1f}%)")
    ax.axis('off')
    plt.colorbar(im, ax=ax, fraction=0.046)

plt.tight_layout()
plt.show()

# Interpretacion Olivetti:
# - PC1 captura la iluminacion global (caras claras vs oscuras)
# - PC2 suele capturar la orientacion lateral (mirando izquierda vs derecha)
# - PC3 puede capturar diferencias en expresion o inclinacion vertical
# - Estas componentes se conocen como "eigenfaces" y son la base del reconocimiento facial clasico
```
